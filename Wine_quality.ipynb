Construction du modle

```python
from sklearn.model_selection import train_test_split
```

```python
X=df.drop("quality",axis=1)
y=df["quality"]

seed=20
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=seed)
print(X_train.shape,y_train.shape)
print(X_test.shape,y_test.shape)
```

```python
sc=StandardScaler()
X_train_sc=sc.fit_transform(X_train)
X_test_sc=sc.transform(X_test)
```

```python
X_train_sc
```

```python
lr_cve= cross_val_score(LinearRegression(),X_train_sc,y_train,scoring="neg_mean_squared_error",cv=3)
rf_cve = cross_val_score(RandomForestRegressor(random_state=seed),X_train_sc,y_train,scoring="neg_mean_squared_error",cv=3)
svm_cve = cross_val_score(SVR(),X_train_sc,y_train,scoring="neg_mean_squared_error",cv=3)
```

```python
lr_cve
```

```python
rf_cve
```

```python
svm_cve
```

```python
print("RF:",rf_cve.mean(),"SVM:",svm_cve.mean(),"LR:",lr_cve.mean())
```

```python
model=RandomForestRegressor(random_state=seed)
model.fit(X_train_sc,y_train)
```

```python
y_pred=model.predict(X_test_sc)
y_pred
```

```python
mean_squared_error(y_test,y_pred)
```

```python
model.feature_importances_
```

```python
vars_imp=pd.Series(model.feature_importances_,index=X.columns)
vars_imp.sort_values(ascending=False)
```

```python
sns.barplot(x=vars_imp,y=vars_imp.index);
plt.title("Importance de chaque caractristique ",fontweight="bold",fontsize=20);
plt.show()
```
